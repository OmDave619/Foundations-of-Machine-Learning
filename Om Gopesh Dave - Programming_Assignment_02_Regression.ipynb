```python
#Understanding Error Surface
#All imports
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import collections

########################################
#Generate meaningfull data
########################################
def wt_ht_relation(heights):
    return [3.86*ht - 110.42 for ht in heights]

def Generate_Data(n,wt_ht_relation,noise_mean,noise_var):

    #generating n radnom data pts for height between 50-76 inches
    heights = np.random.uniform(50,76,n)

    #generating noise
    noise = np.random.normal(noise_mean,noise_var,n)

    #calculating wt according to the given relationship
    weights = wt_ht_relation(heights)

    #adding noise to the calculated wts
    weights = weights + noise

    return heights,weights

########################################
#Plot scatter plot of data
########################################
def scatter_plot_ht_wt(heights,weights):
    plt.figure(figsize=(7,6))
    plt.scatter(heights,weights)
    plt.xlabel('Heights (in inches)')
    plt.ylabel('Weights (in pounds)')
    plt.title(f'Heights vs Weights Scatter plot')
    plt.grid()
    plt.show()

Heights, Weights = Generate_Data(25,wt_ht_relation,0,np.sqrt(20))
scatter_plot_ht_wt(Heights,Weights)
```

```python
########################################
#Weight estimation through error surface, i.e., empirically locate the minima of error surface
########################################
#Sample a bunch of w's around w_opt and compute the associated error

#given w = 3.86h - 110.42
#comparing with, y(x) = w0 + w1x
#we get w_actual = [w0 w1] = [-110.42 3.86]
 
w_actual = np.array([-110.42,3.86])

#sampling 10000 pts around w_optimum 
w0_s = np.linspace(-160,-60,100)
w1_s = np.linspace(1.86,5.86,100)

#creates all possible combination of 100x and 100y , thus 10000 coordinates
w0_s , w1_s = np.meshgrid(w0_s,w1_s)

#flattening the coords
w0_s = w0_s.ravel() 
w1_s = w1_s.ravel()

w_samples = np.vstack((w0_s,w1_s))
w_samples = w_samples.T
w_samples.shape
# temp = np.array([3,4])
# print(np.dot(temp,w[0]))
```

```python
#Compute the error
def Error(w_samples,t,x): #inputs : 1)weight 2)data i.e (t,x)
    
    #w is 1000 by 2 matrix 
    #t is n by 1 vector containing actual targets
    #x is n by 1 vector containing actual inputs/heights
    #we want to return 1000 by 1 vector containing J(w0,w1) for different (w0,w1) in w
 
    #Error vector store the errors for different w0,w1
    error = []

    # making x matrix of type [1 x1
    #                         1 x2] ... (N * (D+1)) , where D is the dimension of input
    ones = np.ones((len(x),1))
    # print(ones.shape,x.shape)
    x_new = np.hstack((ones,x.reshape((len(x),1))))

    #computing error-function J(w) value of each (w0,w1)
    for w in w_samples:

        #Estimate the target
        t_estimated = np.dot(x_new,w)

        #computing the errors
        e = t - t_estimated

        #computing the least square cost function 
        # J(w) = 1/2 (e1^2 + e2^2 + ... en^2)

        J = 1/2 * np.dot(e.T,e)

        error.append(J)

    #Compute and return the error
    return np.array(error)


#Plot 3D error surface and the corresponding contour plots
Errors = Error(w_samples,Weights,Heights)

#reshaping into meshgrid form for plotting
w_0s_mesh = w_samples[:,0].reshape((100,100))
w_1s_mesh = w_samples[:,1].reshape((100,100))
Errors_mesh = Errors.reshape(100,100)    


from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111,projection='3d')
plt.title('Error Surface')
ax.plot_surface(w_0s_mesh,w_1s_mesh,Errors_mesh,cmap = 'viridis',edgecolor='none')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('J(w)')

#Locate the minima of the error surface
min_ind = np.argmin(Errors)
w0_minima = w0_s[min_ind]
w1_minima = w1_s[min_ind]
w_error_surf = np.array([w0_minima,w1_minima])
print(f'Minumum Error is at w0 = {w_error_surf[0]} and w1 = {w_error_surf[1]}')
```

```python

########################################
#Least squares approach to estimate the weights
########################################
#Complete the below linear regression function
def LinearRegression(x,t): #inputs : 1)input data i.e (x). 2)target i.e (t)

    # making x matrix of type [1 x1
    #                         1 x2] ... (N * (D+1)) , where D is the dimension of input
    ones = np.ones((len(x),1))
    x = np.hstack((ones,x.reshape((len(x),1))))

    t = t.reshape((len(x),1))

    w_opt = (np.linalg.inv(x.T@x))@(x.T@t)

    return w_opt


#Estimate optimal weight's using "LinearRegression" function
w_opt = LinearRegression(Heights,Weights)
print(f'Optimum wts using Linear regression formula: w0 = {w_opt[0]}, w1 = {w_opt[1]}')

#Estimate the targets using the input x and the estimated weights
def estimate_targets(x,w):
    # making x matrix of type [1 x1
    #                         1 x2] ... (N * (D+1)) , where D is the dimension of input
    ones = np.ones((len(x),1))
    x = np.hstack((ones,x.reshape((len(x),1))))

    # estimated target t^ = Xw
    t_estimated = np.dot(x,w)

    return t_estimated

t_estimated_least_square = estimate_targets(Heights,w_opt)

#Plot the estimated line on top of the above scatter plot

plt.figure(figsize=(7,6))
plt.scatter(Heights,Weights,label='Input Data')
plt.plot(Heights,t_estimated_least_square,label = 'Predicted line using least squares',c='red')
plt.xlabel('Heights (in inches)')
plt.ylabel('Weights (in pounds)')
plt.title(f'Heights vs Weights Scatter plot')
plt.legend()
plt.grid()
plt.show()

```

```python
#Understanding model order and overfitting
########################################
#Generate 20 data points
########################################

n = 10 #number of training/testing data points

def f(x):
    return np.sin(2*np.pi*x)

def generate_pts(n,f):
    x_input = np.random.uniform(0,1,2*n) #xn is from [0,1],generating random 20 pts
    x_input=np.sort(x_input)
    target = f(x_input) + np.random.normal(0,np.sqrt(0.1),2*n) #generating target using function + noise with mean=0 and var=0.1

    #Obtain train and test splits
    X_train , Y_train = x_input[::2] , target[::2]
    X_test , Y_test = x_input[1::2] , target[1::2]

    return X_train , Y_train , X_test , Y_test

#need to make function of plot 
def plot_initial_graph(X_train,Y_train,X_test,Y_test,f):
    plt.figure(figsize=(7,4))
    x = np.linspace(0,1,100)
    plt.plot(x,f(x),label='Sin(2πx)',c='green')
    plt.scatter(np.hstack((X_train,X_test)),np.hstack((Y_train,Y_test)),edgecolors='blue',facecolors='none')
    plt.xlabel('x (input)')
    plt.ylabel('t (target)')
    plt.title('Model: t = sin(2πx) + ε')
    plt.legend()
    plt.grid()
    plt.show()

X_train , Y_train , X_test , Y_test = generate_pts(n,f)
plot_initial_graph(X_train,Y_train,X_test,Y_test,f)
########################################
#Fitting Mth degree polynomial using least squares approach
########################################
#Complete the function

def Kernel_Matrix(X,M):
    n=len(X)
    K = np.zeros((n,M+1)) 
    for i in range(n):
        for j in range(M+1):
            K[i][j] = X[i]**j
    return K


def PolynomialFit(X_train,Y_train,M,lamda): #(training data, trining targets, Model order, Regularization coefficient)
    #Transform the data using polynomial kernel

    #if x is [x1 x2 ... xn]^T
    #Then Kernel tranformation to M dimensions is phi(x1) = [1 x1 x1^2 ... x1^M]
    #Hence, Kernel matrix K would look like : K = [1 x1 x1^2 ... x1^M 
    #                                              1 x2 x2^2 ... x2^M
    #                                                 ...
    #                                              1 xn xn^2 ... xn^M] n by M matrix

    K = Kernel_Matrix(X_train,M)

    # print(X)

    #Find Pseudo inverse solution
    w_opt = (np.linalg.inv(K.T@K + lamda*np.identity(M+1)))@(K.T@Y_train)

    #return the weight vector
    return w_opt


#Complete the function
def PolynomialPred(w_est,X_train,X_test): #(weight,training data, testing data)
    #Estimate the targets for both training and testing data

    #modifying the x matrix to include ones
    # x_train = np.vstack((np.ones(len(X_train)),X_train)).T
    # x_test = np.vstack((np.ones(len(X_train)),X_train)).T

    #getting kernel matrices
    M = len(w_est) - 1 #as W will M+1 by 1 vector
    K_train = Kernel_Matrix(X_train,M)
    K_test = Kernel_Matrix(X_test,M)

    #estimated value : y=Kw
    est_y_train = np.dot(K_train,w_est)
    est_y_test = np.dot(K_test,w_est)

    #Return training and testing predictions
    return est_y_train,est_y_test

#Complete the function
def PolynomialPred_Error(pred_train,pred_test,Y_train,Y_test): #(training predictions, testing predictions, training targets, testing targets)
    #Estimate the targets for both training and testing data

    #calculating errors
    TrainError = Y_train - pred_train
    TestError = Y_test - pred_test

    #Return training and testing error
    return TrainError,TestError

#Iterate through range of M values
M_range=list(range(10))
TrError = []
TeError = []
for M in M_range:
    #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w = PolynomialFit(X_train,Y_train,M,0)

    #Predict training and testing targets
    tr_pred , te_pred = PolynomialPred(w,X_train,X_test)

    #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    tr_err , te_err = PolynomialPred_Error(tr_pred,te_pred,Y_train,Y_test)


    #Store them for plotting (Storing Root mean square error for plotting)
    TrError.append(np.sqrt(np.mean(tr_err**2)))
    TeError.append(np.sqrt(np.mean(te_err**2)))


#Plot training and testing estimates alogwith the original targets
def plot_graph(X_train,Y_train,X_test,Y_test,order,lamda):
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    n = len(X_train)
    w = PolynomialFit(X_train,Y_train,order,lamda)
    tr_pred, te_pred = PolynomialPred(w,X_train,X_test)

    x = np.linspace(0,1,100)
    x_pred,_= PolynomialPred(w,x,x)

    axes[0].plot(x,x_pred,label='Predicted line',c='red')
    axes[0].plot(x,f(x),label='Sin(Pix)',c='green')

    axes[0].scatter(X_train,Y_train,label='Original training targets',edgecolors='blue',facecolors='none')
    axes[0].scatter(X_train,tr_pred,label='Predicted training Targets',edgecolors='red',facecolors='none')
    axes[0].set_ylim((-1.5,1.5))
    axes[0].legend()
    axes[0].set_title(f'Training estimates M = {order} N = {n} λ = {lamda}')

    x = np.linspace(0,1,100)
    x_pred,_= PolynomialPred(w,x,x)

    axes[1].plot(x,x_pred,label='Predicted line',c='red')
    axes[1].plot(x,f(x),label='Sin(Pix)',c='green')
    axes[1].scatter(X_test,Y_test,label='Original testing targets',edgecolors='blue',facecolors='none')
    axes[1].scatter(X_test,te_pred,label='Predicted testing Targets',edgecolors='red',facecolors='none')
    axes[1].set_ylim((-1.5,1.5))
    axes[1].legend()
    axes[1].set_title(f'Testing estimates M = {order} N = {n} λ = {lamda}')

    plt.show()

plot_graph(X_train,Y_train,X_test,Y_test,0,0)
plot_graph(X_train,Y_train,X_test,Y_test,1,0)
plot_graph(X_train,Y_train,X_test,Y_test,3,0)
plot_graph(X_train,Y_train,X_test,Y_test,9,0)
```
